```python


def rango_a_promedio(rango):
    """Convierte un rango en formato [min, max[ a su valor promedio."""
    if isinstance(rango, str):
        # Extraer los números del rango en formato de string como '[0, 0.2['
        valores = rango.strip('[]').split(', ')
        if len(valores) == 2:
            min_val, max_val = float(valores[0]), float(valores[1].replace('[', ''))
            return (min_val + max_val) / 2
    return None

# Aplicar la conversión a las columnas relevantes
df_districts_info['pct_black/hispanic'] = df_districts_info['pct_black/hispanic'].apply(rango_a_promedio)
df_districts_info['pct_free/reduced'] = df_districts_info['pct_free/reduced'].apply(rango_a_promedio)
df_districts_info['county_connections_ratio'] = df_districts_info['county_connections_ratio'].apply(rango_a_promedio)
df_districts_info['pp_total_raw'] = df_districts_info['pp_total_raw'].apply(rango_a_promedio)

# Verificar la conversión
print(df_districts_info[['pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw']].head())
```

```python
# Eliminar filas con valores nulos en las columnas de interés
df_districts_info_cleaned = df_districts_info.dropna(subset=['pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw'])

# Verificar si hay nulos en las columnas seleccionadas
print(df_districts_info_cleaned[['pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw']].isnull().sum())
```

```python
df_train = df_train.fillna(df_train.median())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, precision_score, f1_score

# Supongamos que 'df_train' es tu DataFrame y la variable a predecir es 'Survived'
X = df_train.drop('Survived', axis=1)  # Características (eliminamos 'Survived')
y = df_train['Survived']  # Variable a predecir

# Dividir el conjunto de datos en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar el modelo de regresión logística
model_logistic = LogisticRegression(max_iter=1000)

# Entrenar el modelo
model_logistic.fit(X_train, y_train)

# Predecir sobre el conjunto de prueba
y_pred = model_logistic.predict(X_test)

# Calcular la precisión (precision)
precision = precision_score(y_test, y_pred)

# Calcular el F1 score
f1 = f1_score(y_test, y_pred)

# Calcular la exactitud (accuracy)
accuracy = accuracy_score(y_test, y_pred)

# Mostrar los resultados
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'F1 Score: {f1:.4f}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar el modelo de árbol de decisión
model_tree = DecisionTreeClassifier(random_state=42)

# Entrenar el modelo
model_tree.fit(X_train, y_train)

#Predecir
y_pred=model_tree.predict(X_test)

# Calcular la precisión (precision)
precision = precision_score(y_test, y_pred)

# Calcular el F1 score
f1 = f1_score(y_test, y_pred)

# Calcular la exactitud (accuracy)
accuracy = accuracy_score(y_test, y_pred)

# Mostrar los resultados
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'F1 Score: {f1:.4f}')
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar el modelo SVM con un kernel lineal
model_svc = SVC(kernel='linear')

# Entrenar el modelo
model_svc.fit(X_train, y_train)

# Predecir sobre el conjunto de prueba
y_pred = model_svc.predict(X_test)

# Calcular la precisión (precision)
precision = precision_score(y_test, y_pred)

# Calcular el F1 score
f1 = f1_score(y_test, y_pred)

# Calcular la exactitud (accuracy)
accuracy = accuracy_score(y_test, y_pred)

# Mostrar los resultados
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'F1 Score: {f1:.4f}')
```
